```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
data=pd.read_csv("GroceryStoreDataSet.csv",header=None)
data.columns=["item1","item2","item3","item4"]
data.drop(columns=["item4"],inplace=True)
```

```python
data
```

```python
li=[]
for i in data.values:
    for j in i:
        li.append(j)
li=list(set(li))
len(li) #11 types of itmes
```

```python
data=pd.get_dummies(data)
data
```

```python
#frequent_pattern
```

```python
frequent_pattern=fpgrowth(data,min_support=0.05,use_colnames=True)
rules=association_rules(frequent_pattern)
```

```python
rules.head()
```

```python
new_rules=np.array(rules.values)
```

```python
for i in new_rules:
    pair=i[0]
    items=[x for x in pair]
    for j in items:
        print(j,",",end=" ")
    print("=>", end=" ")
    pair=i[1]
    items=[x for x in pair]
    for j in items:
        print(j,",",end=" ")
    
    print("Support:",i[4])
    print("Confidence:",i[5])
    print("Lift:",i[6])
    print("************************************")
```
